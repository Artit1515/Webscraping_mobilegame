```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Url = "https://vgsales.fandom.com/wiki/List_of_highest-grossing_mobile_games"
```

```python
page = requests.get(Url)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
table1 = soup.find_all('table',class_ = "wikitable")[0]
table2 = soup.find_all('table',class_ = "wikitable")[1]
print(table1)
```

```python
titles = table1.find_all('th')
print(titles)
```

```python
all_titles = [list_titles.get_text().strip() for list_titles in titles]
print(all_titles)
```

```python
df = pd.DataFrame(columns = all_titles)
df
```
